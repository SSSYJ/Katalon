### Clean up test scripts by removing imports and comments

```python
def script_prepare(script):
    cleaned = []
    lines = script.split('\n')
    for line in lines:
        if line.strip() == '':
            continue
        if line.startswith('import'):
            continue
        if line.strip().startswith('WebUI.comment'):
            continue
        cleaned.append(line)
    return '\n'.join(cleaned)
```

### Get WebUI function with objects as arguments

```python
# Get fun_obj_dict
def update_func_obj_dict(script, func_obj_dict):
    lines = script.split('\n')
    func_name_pattern = re.compile('WebUI\.([a-zA-Z]*)(.*)')
    obj_pattern = re.compile('[^\']*findTestObject*\((\'([^\']*)\')+.*')
    for line in lines:
        result = func_name_pattern.search(line)
        if result:
            func_name = result.group(1)
            obj_result = obj_pattern.search(result.group(2))
            if obj_result:
                obj = obj_result.group(2)
                obj = '/'.join(obj.split('/')[-2:])
                if func_name in func_obj_dict:
                    func_obj_dict[func_name].add(obj)
                else:
                    obj_set = set()
                    obj_set.add(obj)
                    func_obj_dict[func_name] = obj_set
        else:
            continue

    return func_obj_dict
```

```python
def get_func_obj_dict(df):
    func_obj_dict = {}
    for script in df['test_scripts']:
        if script is np.NaN:
            continue
        func_obj_dict = update_func_obj_dict(script, func_obj_dict)
    return func_obj_dict
```

### One line code to dict

```python
def code_to_dict(line):
    d = {}

    var_pattern = re.compile('\t*(def|int|List|String|WebDriver|WebElement|JavascriptExecutor) (\w+) = (.*)')
#     obj_pattern = re.compile('findTestObject\(\'([\w\/]*)\'\)')
    func_name_pattern = re.compile('WebUI\.([a-zA-Z]*)\((.*)\)')
    entity_pattern = re.compile('[a-zA-Z0-9"]+[^,]*|\[[^\]]*\]')

    var_result = var_pattern.search(line)
    # Case 1: TYPE VAR =
    if var_result:
        var_type = var_result.group(1)
        var_name = var_result.group(2)
        result = func_name_pattern.search(var_result.group(3))
        d['var_type'] = var_type
        d['var_name'] = var_name
        if not result:
            d['value'] = var_result.group(3)

    # Case 2: WebUI.FUNCTION
    else:
        result = func_name_pattern.search(line)

    if result:
        func_name = result.group(1)
        entities = entity_pattern.findall(result.group(2))
#         entities = [obj_pattern.search(ent).group(1) if obj_pattern.search(ent) else ent for ent in entities]

        d['action'] = func_name
        d['entities'] = entities
    return d
```

### Scripts process
#### Combine code blocks or multiple code lines separated by comma

```python
def combine_lines(script):
    multi_pattern = "(.*),\n(.*)"
    multi_result = re.search(multi_pattern, script)
    if multi_result:
        sub = ', '.join([part.strip() for part in multi_result.groups()])
        script = re.sub(multi_pattern, sub, script)

    if_block_pattern = "if\s*\(.*\)\s*{\n([^}]*\n)*\t*}"
    if_block_results = re.finditer(if_block_pattern, script)
    for block in if_block_results:
        sub = ''.join(block.group(0).split('\n'))
        script = script.replace(block.group(0), sub)

    return script
```

```python
def code_block_clean(script):
    # Clear up SetUp def
    setup_pattern = "@SetUp\(\)\ndef setUp\([^\)]*\) {([^}]*)}"
    setup_result = re.search(setup_pattern, script)
    if setup_result:
        sub = setup_result.group(1)
        script = re.sub(setup_pattern, sub, script)

    # Clear up if(true)
    if_true_pattern = "if\s*\(true\)\s*{([^}]*)}"
    if_true_result = re.finditer(if_true_pattern, script)
    for if_true_block in if_true_result:
        sub = '\n'.join(if_true_block.group(1).split('\t'))
        script = script.replace(if_true_block.group(0), sub)

    return script
```

```python
def if_block_to_dict(block_line):
    '''Process if blocks to dict'''
    d = {}
    condition_pattern = "if\s*\((.*)\)\s*{(.*)}$"
    condition_result = re.search(condition_pattern, block_line)
    d['condition'] = condition_result.group(1)
    actions = condition_result.group(2).split('\t')
    actions_dict = []
    for action in actions:
        if action.strip() == '':
            continue
        actions_dict.append(code_to_dict(action))
    d['actions'] = actions_dict
    return d
```

```python
def scripts_preprocess(df):
    '''Process all test scripts'''

    processed_scripts = []
    for old_script in df['prepared_scripts']:
        if old_script is np.NaN:
            continue

        script = code_block_clean(combine_lines(old_script))

        lines = script.split('\n')
        new_script = []
        block = []
        for line in lines:
            line = line.strip()
            # ignore empty lines
            if line == '':
                continue
            # ignore comments
            if line.startswith('\'') and line.endswith('\''):
                continue

            # process WebUI functions and def VAR = code lines
            if line.startswith('WebUI.') or ('=' in line):
                if len(block) > 0:
                    new_script.append('\n'.join(block))
                    block = []
                new_line = code_to_dict(line)
                new_script.append(new_line)

            # process If blocks
            elif line.startswith('if'):
                new_line = if_block_to_dict(line)
                new_script.append(new_line)

            else:
                block.append(line)
        processed_scripts.append(new_script)
#         for line in new_script:
#             print(line)
#         print('\n')
    return processed_scripts
```

### Nested function names

```python
# Get all nested functions
def get_nested_function(line, nested_function_set):
    outside_pattern = "WebUI.[a-zA-Z]+\((.*)\)"
    results = re.findall(outside_pattern, line)

    nested_patterns = "([\w.]+)\(.*\)"
    if results:
        for result in results:
            nested_result = re.search(nested_patterns, result)
            if nested_result:
                nested_function_set.update(nested_result.groups())
    return nested_function_set
```

### Revert back to scripts

```python
func_verb_alignment = {
    'setText': {'enter', 'input'},
    'click': {'click'},
    'waitForElementPresent': {'wait'},
    'getText': {'get'},
    'verifyMatch': {'check'},
    ''
}

verb_func_alignment = {}
for func, verbs in func_verb_alignment.items():
    for verb in verbs:
        verb_func_alignment[verb] = func
verb_func_alignment
```

```python
import Levenshtein
from collections import defaultdict

path_dict = defaultdict(lambda: defaultdict(set))
for values in func_obj_dict.values():
    for value in values:
        get_path_list = value.split('/')
        path = '/'.join(get_path_list[0:-1])
        get_element_list = get_path_list[-1].split('_')
        element = get_element_list[1:]
        if len(get_element_list) > 1:
            element = '_'.join(get_element_list[1:])
            path_dict[path][element].add(value)
        else:
            element = get_element_list[0]
            path_dict[path][element].add(value)

action_widget = {
    'input': {'tbx', 'inp', 'input', 'textarea', 'ipn', 'ta', 'ckeditor', 'ckedior', 'updateOption', 'deleteOption'},
    'enter': {'tbx', 'inp', 'input', 'tbx', 'textarea', 'ipn', 'ta', 'ckeditor', 'ckedior', 'updateOption', 'deleteOption'},
    'click': {'img', 'icon', 'lbl', 'label', 'star', 'cbx', 'btn', 'tab', 'toggle', 'switch'},
    'click_content': {'radio', 'select', 'slt', 'rdi', 'link'},
    'container': {'Span', 'span', 'modal', 'wrapper', 'div', 'dialog', 'nav'},
    'readable': {'successToast', 'th', 'title', 'toast', 'txt', 'tr', 'content', 'cpn', 'table'}
}


def get_page(page):
    min_dist = np.inf
    min_key = ''
    for key in path_dict.keys():
        distance = Levenshtein.distance(page, key.lower())
        if distance < min_dist:
            min_dist = distance
            min_key = key
    return min_key


def get_location(page, location):
    min_dist = np.inf
    min_key = ''
    for key in path_dict[page].keys():
        distance = Levenshtein.distance(location, key.lower())
        if distance < min_dist:
            min_dist = distance
            min_key = key
    return min_key


def get_obj_path(parsed_test_case):
    page = get_page(parsed_test_case['page'])
    if 'location' in parsed_test_case.keys():
        location = get_location(page, parsed_test_case['location'])
    else:
        location = get_location(page, parsed_test_case['value'])
    for element in path_dict[page][location]:
        if element.split('/')[-1].split('_')[0] in action_widget[parsed_test_case['V']]:
#             return element
            return f'findTestObject(\'{element}\')'
        else:
            return None
```

```python
def align_one_line_to_script(test_case_dict):
    action = test_case_dict['V']
    fun = ''
    args = []

    fun = verb_func_alignment[action]
    if fun == 'setText':
        arg1 = get_obj_path(test_case_dict)
        arg2 = f"'{test_case_dict['value']}'"
        args.extend([arg1, arg2])
    if fun == 'click':
        arg1 = get_obj_path(test_case_dict)
        args.append(arg1)
    if fun == 'waitForElementPresent':
#         arg1 = get_obj_path(test_case_dict)
        args.extend([test_case_dict['page'], '30'])
    if fun == 'getText':
#         arg1 = get_obj_path(test_case_dict)
        args.append(test_case_dict['page'])
    if fun == 'verifyMatch':
        args.extend([test_case_dict['page'], test_case_dict['page'], 'false'])

    return {"action": fun, "entities": args}


def revert_to_script_dicts(test_case):
    script_dicts = []
    for line_dict in test_case:
        if line_dict['V'] not in verb_func_alignment:
            continue
        script_dict = align_one_line_to_script(line_dict)
        script_dicts.append(script_dict)
    return script_dicts


def revert_to_script(script_dicts):
    script = []
    for line_dict in script_dicts:
        line = f"WebUI.{line_dict['action']}({', '.join(line_dict['entities'])})"
        script.append(line)
    return '\n'.join(script)
```
